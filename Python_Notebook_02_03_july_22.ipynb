```python
#import
import numpy as np
import matplotlib.pyplot as plt
#data
data = {"English":89, "Math":67, "Science":90,"history":70}
#
subject = list(data.keys())
Marks = list(data.values())
#size od plot
fig = plt.figure(figsize=(8,2))

plt.bar(subject,Marks , color = ['g','b','r','y'])

plt.title("This is my bar graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

```python
import matplotlib.pyplot as plt

lab = 'English', 'Math', 'Science','History'
sizes = [85,89,75,90]

exp = (0,0.5,0.1,0.2)

fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=exp, labels=lab, autopct='%1.3f%%',
        shadow=True, startangle=0)

plt.show()
```
